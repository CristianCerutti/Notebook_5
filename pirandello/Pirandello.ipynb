```python
file = open("sei_personaggi_in_cerca_dautore_Luigi_Pirandello.txt").read()
file = file.replace("\n", " ")
file = file.split(" ")
diz = {}
diz2 = {}
for parola in file:
    if parola == "":
        continue
    if parola not in diz:
        diz[parola] = 1
    else:
        diz[parola] += 1
print(diz)
```
