Creare una matrice fatta come un bersaglio 10x10, le cornici concentriche hanno valori crescenti verso il centro (quella più esterna ha valori = 1)

```python
Z = np.ones((10,10))

for i in range (10//2-1,-1,-1):
    Z[i,:] = i+1
    Z[10-i-1,:] = i+1
    Z[:,i] = i+1
    Z[:,10-i-1] = i+1
print(Z)
```

Creare una funzione che trovi gli elementi comuni a due ndarray monodimensionali di lunghezza diversa.

```python
def trova_comuni(array1, array2):
    lista = []
    if array1.size > array2.size:
        for elemento in array1:
            if elemento in array2:
                lista.append(elemento)
    else:
        for elemento in array2:
            if elemento in array1:
                lista.append(elemento)
    array3 = np.array(lista)
    return array3

arrayA = np.array([1,2,3,4,5,10,9,12])
arrayB = np.array([1,2,7,8,3,10])
print(trova_comuni(arrayA,arrayB))
```
